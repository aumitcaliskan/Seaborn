```python
planets = sns.load_dataset("planets")
```

```python
planets
```

```python
cmap = sns.cubehelix_palette(rot=-.2, as_cmap=True)
g = sns.relplot(data=planets, x="distance", y="orbital_period", hue="year", size="mass", palette=cmap, sizes=(10, 200))
g.set(xscale="log", yscale="log")
g.ax.xaxis.grid(True, "minor", linewidth=.25)
g.ax.yaxis.grid(True, "minor", linewidth=.25)
g.despine(left=True, bottom=True)
```

```python
flights = sns.load_dataset("flights")
```

```python
flights
```

```python
g = sns.relplot(data=flights, x="month", y="passengers", col="year", hue="year", kind="line",
                palette="crest", linewidth=4, zorder=5, col_wrap=3, height=2, aspect=1.5, legend=False)
for year, ax in g.axes_dict.items():
    ax.text(.8, .85, year, transform=ax.transAxes, fontweight="bold")
    
sns.lineplot(data=flights, x="month", y="passengers", units="year", estimator=None, color=".7", linewidth=1, ax=ax)

ax.set_xticks(ax.get_xticks()[::2])

g.set_titles("")
g.set_axis_labels("", "Passengers")
g.tight_layout()
```

```python
sns.set_theme(style="ticks")
```

```python
f, ax = plt.subplots(figsize=(7, 6))
ax.set_xscale("log")
sns.boxplot(x="distance", y="method", data=planets, whis=[0, 100], width=.6, palette="vlag")
sns.stripplot(x="distance", y="method", data=planets, size=4, color=".3", linewidth=0)
ax.xaxis.grid(True)
ax.set(ylabel="")
sns.despine(trim=True, left=True)
```
