```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# two continuous features
# numeric variables (age, height, salary, temperature)
# always between two values
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(12,4), dpi=200)

sns.scatterplot(x='salary',y='sales', data=df, hue='level of education')

#hue colors the points according to the column
```

```python
plt.figure(figsize=(12,4), dpi=200)

sns.scatterplot(x='salary',y='sales', data=df, hue='salary')

#hue colors the points according to the column(continuous or categorical)
```

```python
# https://matplotlib.org/tutorials/colors/colormaps.html
```

```python
plt.figure(figsize=(12,4), dpi=200)

sns.scatterplot(x='salary',y='sales', data=df, hue='level of education',
               palette='Dark2')
```

```python
plt.figure(figsize=(12,4), dpi=200)

sns.scatterplot(x='salary',y='sales', data=df, size='work experience')
```

```python
plt.figure(figsize=(12,4), dpi=200)

sns.scatterplot(x='salary',y='sales', data=df, s=100, alpha=0.2, style='level of education', hue='level of education')

# s size of plots
# alpha transparency

plt.savefig('my_plot.jpg')
```
