```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Distrubution within categories: boxplot, violinplot, swarmplot, boxenplot(letter-value plot)

## main one is the boxplot
# boxplot uses quartiles. Quartiles seperate out the data into 4 equal number of data points. 
# Q2 is the median
# 50% of all data points are inside the box. IQR= Q3-Q1. Q1 is the 25th percentile. Q3 is the 75th percentile
# Whiskers 1.5*IQR
# Outside of the whiskers : Outliers 

## violinplot
# KDE of a single feature. Mirror it on x axis. Then combine it to get the violin plot

## swarmplot for very large data sets

## boxenplot very uncommon. designed as an expansion upon the normal box plot
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.boxplot(data=df, y='reading score', x='parental level of education', hue='test preparation course')
plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
plt.figure(figsize=(10,8),dpi=200)
sns.boxplot(data=df, x='reading score', y='parental level of education', hue='test preparation course', palette='Set3')
plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
plt.figure(figsize=(10,8),dpi=200)

sns.violinplot(data=df, x='reading score', y='parental level of education')

# using hue is better on boxplot
```

```python
plt.figure(figsize=(10,8),dpi=200)

sns.violinplot(data=df, x='reading score', y='parental level of education', hue='test preparation course', palette='Set2',
              split=False, inner='stick', bw=0.5)

# split=True combines violins according to the hue.
# inner=None removes the white dot
# inner='quartiles'
# bw kde

plt.legend(bbox_to_anchor=(1.25,0.5))

# using hue is better on boxplot
```

```python
sns.swarmplot(data=df, x='math score')
```

```python
plt.figure(figsize=(8,4),dpi=200)
sns.swarmplot(data=df, x='math score',y='gender',size=2,
             hue='test preparation course', dodge=True)

plt.legend(bbox_to_anchor=(1.25,1))
```

```python
sns.boxenplot(x='math score', y='test preparation course', data=df, hue='gender')
```

```python
sns.boxplot(x='math score', y='test preparation course', data=df, hue='gender')
```
