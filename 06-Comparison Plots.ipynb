```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2 dimensional. jointplot and pairplot
# jointplot : histograms with scatterplot or 2-dimensional KDE plots show shaded distribution
# pairplot : quick way to compare all "numerical" columns,
# histogram for each column and a scatterplot comparison between columns
```

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

```python
sns.jointplot(data=df, x='math score', y='reading score')
```

```python
sns.jointplot(data=df, x='math score', y='reading score', kind='hex')
```

```python
sns.jointplot(data=df, x='math score', y='reading score', kind='scatter', alpha=0.2)
```

```python
sns.jointplot(data=df, x='math score', y='reading score', kind='hist')
```

```python
sns.jointplot(data=df, x='math score', y='reading score', kind='kde',shade=True)
```

```python
sns.jointplot(data=df, x='math score', y='reading score',hue='gender')
```

```python
sns.pairplot(data=df)

# numerical columns
```

```python
sns.pairplot(data=df,hue='gender')

# default KDE

# be careful about running all dataframe. It can take time.
```

```python
sns.pairplot(data=df,hue='gender',diag_kind='hist')

# hard to read histogram
```

```python
sns.pairplot(data=df,hue='gender', corner=True)

# delete the same plots
```
