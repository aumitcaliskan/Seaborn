```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# visual equivalent of displaying a pivot table. Numeric values.

# heatmap() : distrubition of cell values with a color mapping. All the cells should be the same unit.

# clustermap() : first conducts hierarchical clustering to reorganize data into groups
```

```python
df = pd.read_csv('country_table.csv')
```

```python
df = df.set_index("Countries")

# perfect for heatmap
```

```python
df
```

```python
sns.heatmap(df)

#life expectancy is different unit(year ) than the others(rate)
```

```python
plt.figure(dpi=200)
sns.heatmap(df.drop('Life expectancy',axis=1), linewidth=0.5, annot=True, cmap='viridis')

#choose the correct colors for colorblind people
```

```python
plt.figure(dpi=200)
sns.heatmap(df.drop('Life expectancy',axis=1), linewidth=0.5, annot=True, cmap='viridis', center=40)

# center changes the colormaping. No need to change. Default one is good.
```

```python
plt.figure(dpi=200)
sns.clustermap(df.drop('Life expectancy',axis=1), linewidth=0.5, annot=True, cmap='viridis')
```

```python
plt.figure(dpi=200)
sns.clustermap(df.drop('Life expectancy',axis=1), linewidth=0.5, annot=True, cmap='viridis',
              col_cluster=False)
```
